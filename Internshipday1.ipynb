```python
df = df.dropna(axis = 1,how = "all" )  #Removes columns containing all NaN value 
```

```python
from datetime import date
```

```python
df['date_added'] = df['date_added'].astype('datetime64[ns]')
```

```python
df.isnull().sum()  # Identifying number of null values in a column
```

```python
df
```

```python
df['director'] = df['director'].fillna('Not Listed')  #director column contains a lot of values, will hamper our data so we will use fillna and enter Not Listed in nulls
```

```python
df['cast'] = df['cast'].fillna('Not Listed')
```

```python
df['country'] = df['country'].fillna('Not Listed')
```

```python
df['date_added'].isnull().sum()
```

```python
df.dropna(subset = 'date_added',inplace = True)
```

```python
df.dropna(subset = 'rating',inplace = True)
```

```python
df.dropna(subset = 'duration',inplace = True)
```

```python
df['duration'] = df['duration'].str.rstrip("s") # Removes s from right side in duration column
```

```python
df
```

```python
# Create a new column 'duration_minutes'
df['duration_minutes'] = np.where(df['duration'].str.contains('min'), 
                                 df['duration'].str.split().str[0], 
                                 np.nan)
# Create a new column 'duration_seasons'
df['duration_seasons'] = np.where(df['duration'].str.contains('Season'), 
                                 df['duration'].str.split().str[0], 
                                 np.nan)
```

```python
df
```

```python
df['duration_minutes'] = df['duration_minutes'].fillna('0') #Fills 0 where duration is not in minutes only in season
```

```python
df['duration_seasons'] = df['duration_seasons'].fillna('0') # Fills 0 where duration is not in season only in minutes
```
